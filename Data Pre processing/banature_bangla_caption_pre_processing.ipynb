```python
import pandas as pd
import json

# Load the CSV file
csv_file_path = '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap.csv'
csv_data = pd.read_csv(csv_file_path)

# Load the JSON files
def load_json_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return json.load(file)

json_file_paths = [
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_test.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_train.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_val.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_test.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_train.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_val.json'
]

# Loading all JSON data
json_data = [load_json_file(file_path) for file_path in json_file_paths]

# Load the text file
txt_file_path = '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/bnature.txt'
with open(txt_file_path, 'r', encoding='utf-8') as txt_file:
    txt_data = txt_file.readlines()

# Visualize the data
print("CSV Data Preview:")
print(csv_data.head())

print("\nSample JSON Data from one of the JSON files:")
print(json_data[0])  # Print first JSON file for visualization

print("\nSample TXT Data:")
print(txt_data[:5])  # Preview first 5 lines

# Combine all datasets into one dictionary
combined_data = {
    'csv_data': csv_data.to_dict(orient='records'),  # Convert CSV to a list of dictionaries
    'json_data': json_data,  # JSON data already in list format
    'txt_data': txt_data  # Keeping raw text data
}

# Save the combined data to a new JSON file
output_file_path = '/content/drive/MyDrive/Bangla Image dataset with caption/combined_data.json'
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    json.dump(combined_data, output_file, ensure_ascii=False, indent=4)

print(f"Combined data saved to {output_file_path}")
```

```python
import pandas as pd
import json

# Load the CSV file
csv_file_path = '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap.csv'
csv_data = pd.read_csv(csv_file_path)

# Load the JSON files
def load_json_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return json.load(file)

json_file_paths = [
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_test.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_train.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/BanCap_val.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_test.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_train.json',
    '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/Bnature_val.json'
]

# Load text file
txt_file_path = '/content/drive/MyDrive/Bangla Image dataset with caption/BAN CAP and BNature/BanglaImageCaptioning/annotations/bnature.txt'
with open(txt_file_path, 'r', encoding='utf-8') as txt_file:
    txt_data = txt_file.readlines()

# Load the JSON files and count captions
json_data = [load_json_file(file_path) for file_path in json_file_paths]

# Function to count captions in JSON files
def count_json_captions(json_data):
    caption_count = 0
    for data in json_data:
        for key, value in data.items():
            # Assuming captions are stored in a key called 'caption' or similar
            if 'caption' in value:
                caption_count += len(value['caption'])
    return caption_count

# Counting captions in the CSV file
# Assuming the captions are stored in a column named 'caption'
csv_caption_count = csv_data['caption'].count() if 'caption' in csv_data.columns else 0

# Counting captions in JSON files
json_caption_count = sum([count_json_captions(data) for data in json_data])

# Counting captions in the text file
txt_caption_count = len(txt_data)  # Each line could be considered a caption

# Total caption count
total_caption_count = csv_caption_count + json_caption_count + txt_caption_count

# Display the results
print(f"CSV Caption Count: {csv_caption_count}")
print(f"JSON Caption Count: {json_caption_count}")
print(f"Text Caption Count: {txt_caption_count}")
print(f"Total Caption Count: {total_caption_count}")
```
